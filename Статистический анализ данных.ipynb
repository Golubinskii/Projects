##Примечание
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.

Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.


Примите, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

В датафрейме `calls` столбец `call_date` имеет тип `object` требуется его преобразование в тип `datetime`.
Столбец `id` содержит номер записи в датафрейме для наших задач этот столбец не понадобится и его можно будет удалить. Также потребуется выделение номера месяца из даты с созданием столбца `month`. Для этих целей напишем функцию `datetime_extract_month`, т.к. такие же преобразования требуются и для датафрейма `internet` и `messages`.

В датафрейме `users` нужно преобразовать столбцы `reg_date` и `churn_date` в тип `datetime`. Столбец `churn_date` содержит пропуски, скорее всего они означают, что на момент выгрузки этих данных пользователь пользовался услугами оператора связи.

```python
def datetime_extract_month(df,column, *column_del):
  df =df.drop(*column_del, axis=1)
  df[column] = pd.to_datetime(df[column], format='%Y-%m-%d')
  df['month'] = pd.DatetimeIndex(df[column]).month
  return df
```

Рассчитаем количество сделанных звонков, а также их общую длительность по месяцам. Согласно правилам тарификации произведем округление длительности звонков с точности до минуты в большую сторону.Далее отфильтруем ненулевые длительности звоноков, т.к. согласно условию это пропущенные звонки и далее сгруппируем данные по `user_id` и `month`.

```python
calls['duration'] = calls['duration'].apply(lambda x: math.ceil(x))
calls
```

Далее реализуем функцию, которая на основе тарифа и установленных лимитов в нем будет расчитывать выручку для каждого клиента.

```python
def calculate_revenue(row):
  if row['tariff'] == 'smart':
    monthly_fee = tariffs.loc[0,'rub_monthly_fee'] # ежемесячная плата за тариф Smart

    amount_calls = row['duration'] - tariffs.loc[0,'minutes_included']
    if amount_calls < 0:
      amount_calls=0  # если абонент не превысил лимит по минутам,
    else:             # то дополнительных средств с него не списывается
        amount_calls = amount_calls*tariffs.loc[0,'rub_per_minute']
        # в случае превышения лимита берется плата за доп. минуты согласно тарифа


    # аналогично для сообщений
    amount_messages = row['number_messages'] - tariffs.loc[0,'messages_included']
    if amount_messages < 0:
      amount_messages=0
    else:
        amount_messages = amount_messages*tariffs.loc[0,'rub_per_message']
    # аналогично для интернета
    amount_internet = row['gb_used'] - tariffs.loc[0,'mb_per_month_included']/1024
    if amount_internet < 0:
      amount_internet=0
    else:
        amount_internet = amount_internet*tariffs.loc[0,'rub_per_gb']

    # аналогично для другого тарифа

  else:

    monthly_fee = tariffs.loc[1,'rub_monthly_fee']

    amount_calls = row['duration'] - tariffs.loc[1,'minutes_included']
    if amount_calls < 0:
      amount_calls=0
    else:
        amount_calls = amount_calls*tariffs.loc[1,'rub_per_minute']

    amount_messages = row['number_messages'] - tariffs.loc[1,'messages_included']
    if amount_messages < 0:
      amount_messages=0
    else:
        amount_messages = amount_messages*tariffs.loc[1,'rub_per_message']

    amount_internet = row['gb_used'] - tariffs.loc[1,'mb_per_month_included']/1024
    if amount_internet < 0:
      amount_internet=0
    else:
        amount_internet = amount_internet*tariffs.loc[1,'rub_per_gb']

  return  monthly_fee +  amount_calls +  amount_messages + amount_internet

```

```python
df['revenue'] = df.apply(calculate_revenue, axis=1)
```

Для подсчета среднего количества минут, дисперсии и стандартного отклонения исключим из расчета нулевые значения. Так мы получим цифры для тех абонентов, которые пользовались данной услугой.

```python
df_smart_calls = df_smart[df_smart['duration']!=0]
df_ultra_calls = df_ultra[df_ultra['duration']!=0]

def mean_var_std(lst_smart, lst_ultra):
  m_s = lst_smart.mean()
  v_s = np.var(lst_smart, ddof=1)
  s_s = v_s**0.5

  m_u = lst_ultra.mean()
  v_u = np.var(lst_ultra, ddof=1)
  s_u = v_u**0.5

  mtx = [['Smart', m_s, v_s, s_s ], ['Ultra', m_u, v_u, s_u]]

  d = pd.DataFrame(data=mtx, columns=['Тариф', 'Ср. значение', 'Дисперсия', 'Ст. отклонение'])
  return d
```

```python
result_calls = mean_var_std(df_smart_calls['duration'], df_ultra_calls['duration'])
result_calls
```
