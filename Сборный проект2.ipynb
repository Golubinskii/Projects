**Описание проекта**

Заказчик этого исследования — сеть отелей «Как в гостях».

Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.

Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

**Бизнес-метрика и другие данные**

Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.

В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.

Стоимость номеров отеля:
 - категория A: за ночь — 1 000, разовое обслуживание — 400;
 - категория B: за ночь — 800, разовое обслуживание — 350;
 - категория C: за ночь — 600, разовое обслуживание — 350;
 - категория D: за ночь — 550, разовое обслуживание — 150;
 - категория E: за ночь — 500, разовое обслуживание — 150;
 - категория F: за ночь — 450, разовое обслуживание — 150;
 - категория G: за ночь — 350, разовое обслуживание — 150.

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.

Убытки отеля в случае отмены брони номера — это стоимость одной уборки и одной ночи с учётом сезонного коэффициента.

На разработку системы прогнозирования заложен бюджет — 400 000. При этом необходимо учесть, что внедрение модели должно окупиться за тестовый период. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

##**Шаг 3. Вычислите бизнес-метрику**
**Оцените прибыль отеля без внедрения депозитов.**

Составим словари для вычисления прибыли и функцию расчета количества уборок от общего количества ночей по принципу: 1 ночь - 1 уборка, 2 ночи - 2 уборки, 3 ночи - 2 уборки, 4 ночи - 2 уборки, 5 ночей - 3 уборки, 6 - ночей 3 уборки и т.д.

```python
d_room = {'A':[1000, 400],   # ключ - категория номера, значение
          'B':[800, 350],    # списке с индексом 0 - цена за проживание,
          'C':[600, 350],    # значение в списке с индексом 1 - цена за
          'D':[550, 150],    # за уборку
          'E':[500, 150],
          'F':[450, 150],
          'G':[350, 150]
          }
d_season = {'January': 1,    # повышающие сезонные коеэффициенты
            'February': 1,
            'March': 1.2,
            'April': 1.2,
            'May' : 1.2,
            'June' : 1.4,
            'July' : 1.4,
            'August' : 1.4,
            'September': 1.2,
            'October' : 1.2,
            'November' : 1.2,
            'December' : 1
            }
def number_cleaning(n):
  return n//2 + n%2
```

```python
train.columns
```

Напишем функцию для расчета прибыли и убытка. Если клиент не отменял бронь, то отель получает прибыль равную стоимости номера соответсвующей категории умноженной на сезонный коэффициент и также умноженный на общее количество ночей за минусом стоимости уборки.
Если клиент все же отменил бронь, то отель получает убыток в размере стоимости одной уборки и одной ночи с учетом сезонного коэффициента.

```python
def profit(row):
  season_coeff = d_season[row['arrival_date_month']]
  if row['is_canceled'] == 1:
    accomodation = season_coeff*d_room[row['reserved_room_type']][0]
    cleaning = d_room[row['reserved_room_type']][1]
    return - (accomodation + cleaning)

  if row['is_canceled'] == 0:
    accomodation = row['total_nights']*d_room[row['reserved_room_type']][0]*season_coeff
    cleaning = number_cleaning(row['total_nights'])*d_room[row['reserved_room_type']][1]
    return accomodation - cleaning
```

Напишем функцию для расчета прибыли и убытков с учетом предсказания модели:
- если модель угадывала и клиент приезжал, отель получает прибыль равную произведению количества ночей умноженной на сезонный коэффициент за вычетом количества уборок умноженного на стоимость одной уборки;
- если модель угадывала и клиент не приезжал отель получал прибыль за одну ночь равную 80% от стоимости номера с учетом сезонного коэффициента плюс стоимость одной уборки;
- если же модель допускала ошибки 1-го или 2-го рода, то отель получал убыток в размере стоимости одной ночи с учетом сезонного коэффициента плюс стоимости одной уборки.

```python
def profit_deposit(row):
  season_coeff = d_season[row['arrival_date_month']]
  if (row['is_canceled'] == 0 and row['predictions'] == 0):
    accomodation = row['total_nights']*d_room[row['reserved_room_type']][0]*season_coeff
    cleaning = number_cleaning(row['total_nights'])*d_room[row['reserved_room_type']][1]
    return accomodation - cleaning

  if row['is_canceled'] == 1 and row['predictions'] == 1:
    accomodation = 0.8*d_room[row['reserved_room_type']][0]*season_coeff
    cleaning =  d_room[row['reserved_room_type']][1]
    return accomodation + cleaning

  if (row['is_canceled'] == 1 and row['predictions'] == 0) or (
      row['is_canceled'] == 0 and row['predictions'] == 1):
    accomodation = season_coeff*d_room[row['reserved_room_type']][0]
    cleaning = d_room[row['reserved_room_type']][1]
    return - (accomodation + cleaning)
```
